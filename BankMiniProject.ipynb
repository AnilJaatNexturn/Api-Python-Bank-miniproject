```python
columns = ['central_bank', 'country', 'rate_pct', 'last_updated']
with open('centralbank.csv', 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=columns)
    writer.writeheader()
    for item in data['central_bank_rates']:
        writer.writerow({column: item[column] for column in columns})
```

```python
columns = ['name', 'rate_pct', 'last_updated']
with open('noncentralbank.csv', 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=columns)
    writer.writeheader()
    for item in data['non_central_bank_rates']:
        writer.writerow({column: item[column] for column in columns})
```
